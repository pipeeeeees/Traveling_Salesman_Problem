```python
%matplotlib inline
import numpy as np
import math
import matplotlib.pyplot as plt
import copy
import random
import csv
import time
from mpl_toolkits.basemap import Basemap
```

```python
def distancelength(x1,y1,x2,y2):
    return np.linalg.norm(np.array([(x1 - x2)*54.6, (y1 - y2)*69]))

def NNplotupdate():
    ax.clear()
    ax.plot(finalX,finalY)
    ax.plot(allthexpoints,alltheypoints,'o')
    ax.set_title('Nearest Neighbor')
    ax.set_xlabel('X-Axis (meters)')
    ax.set_ylabel('Y-Axis (meters)')
    fig.canvas.draw()
    plt.pause(0.0000001)

def optplotupdate():
    ax.clear()
    ax.plot(twooptX,twooptY)
    ax.plot(finalX,finalY)
    ax.plot(allthexpoints,alltheypoints,'.')
    ax.set_title('2-opt')
    ax.set_xlabel('X-Axis (meters)')
    ax.set_ylabel('Y-Axis (meters)')
    fig.canvas.draw()
    plt.pause(0.0000001)

def pathlength(xpoints,ypoints):
    optdistance = 0
    for i in range(len(xpoints)-1):
        optdistance += distancelength(xpoints[i+1],ypoints[i+1],xpoints[i],ypoints[i])
    return optdistance
```

```python
allthexpoints = []
alltheypoints = []
caps = []

with open('us-state-capitals.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            print(f'\t{row[0]} \t {row[1]} \t {row[2]}\t{row[3]}')
            if (float(row[3]) > -150):
                caps.append(row[1])
                allthexpoints.append(round(float(row[2]),3))
                alltheypoints.append(round(float(row[3]),1))
            line_count += 1
    print(f'Processed {line_count} lines.')
```

```python
fig = plt.figure(figsize=(12, 12))
m = Basemap(projection='lcc', resolution='l',
            width=8E6, height=8E6, 
            lat_0=45, lon_0=-100,)
m.etopo(scale=0.5, alpha=0.5)
m.drawcoastlines(color='gray')
m.drawcountries(color='gray')
m.drawstates(color='gray')

for i in range(len(caps)):
    # Map (long, lat) to (x, y) for plotting
    x, y = m(alltheypoints[i], allthexpoints[i])
    plt.plot(x, y, 'ok', markersize=5)
    plt.text(x, y, ' ' + caps[i], fontsize=10);
```

```python
fig = plt.figure(figsize=(12, 12), edgecolor='w')
m = Basemap(projection='cyl', resolution='l',
            llcrnrlat=20, urcrnrlat=70,
            llcrnrlon=-140, urcrnrlon=-58, )
m.etopo(scale=0.5, alpha=0.5)
m.drawcoastlines(color='gray')
m.drawcountries(color='gray')
m.drawstates(color='gray')
for i in range(len(caps)):
    # Map (long, lat) to (x, y) for plotting
    x, y = m(alltheypoints[i], allthexpoints[i])
    plt.plot(x, y, 'ok', markersize=5)
    plt.text(x, y, ' ' + caps[i], fontsize=9);
```

```python
finalX = []
finalY = []
alltheXXpoints = copy.copy(allthexpoints)
alltheYYpoints = copy.copy(alltheypoints)
tempXval = -134
tempYval = 58
for i in range(len(allthexpoints)): #to cover all the points in the array
    distance = 100000000 #arbitrary long distance
    for j in range(len(alltheXXpoints)): #to cover all the unvisited points
        if distancelength(tempXval,tempYval,alltheXXpoints[j],alltheYYpoints[j]) < distance:
            distance = distancelength(tempXval,tempYval,alltheXXpoints[j],alltheYYpoints[j])
            indexsaver = j
    tempXval = alltheXXpoints[indexsaver]
    tempYval = alltheYYpoints[indexsaver]
    finalX.append(alltheXXpoints[indexsaver])
    finalY.append(alltheYYpoints[indexsaver])
    alltheXXpoints.pop(indexsaver)
    alltheYYpoints.pop(indexsaver)
    #NNplotupdate()
finalX.append(float(finalX[0]))
finalY.append(float(finalY[0]))
#NNplotupdate()
print("NN Produced a path length of {} Miles".format(round(pathlength(finalX,finalY),2)))

fig = plt.figure(figsize=(12, 12), edgecolor='w')
m = Basemap(projection='cyl', resolution='l',
            llcrnrlat=20, urcrnrlat=70,
            llcrnrlon=-140, urcrnrlon=-58, )
m.etopo(scale=0.5, alpha=0.5)
m.drawcoastlines(color='gray')
m.drawcountries(color='gray')
m.drawstates(color='gray')
for i in range(len(caps)):
    # Map (long, lat) to (x, y) for plotting
    x, y = m(alltheypoints[i], allthexpoints[i])
    plt.plot(x, y, 'ok', markersize=5)
    plt.text(x, y, ' ' + caps[i], fontsize=7);
plt.plot(finalY,finalX)
print("NN Produced a path length of {} Miles".format(round(pathlength(finalY,finalX),2)))
```

```python
twooptX = copy.copy(finalX)
twooptY = copy.copy(finalY)
oldX = 0
oldY = 0
shrinkingdistance = pathlength(twooptX,twooptY)
optswaplength = 10
print("Starting the 2-opt point swap...")
for i in range(len(twooptX)):
    for rand in reversed(range(0,15)):
        #rand = random.randint(1,3)
        oldX = twooptX[(i+(rand))%len(twooptX)]
        oldY = twooptY[(i+(rand))%len(twooptX)]
        twooptX[(i+(rand))%len(twooptX)] = twooptX[i]
        twooptY[(i+(rand))%len(twooptX)] = twooptY[i]
        twooptX[i] = oldX
        twooptY[i] = oldY
        #optplotupdate()
        optdistance = pathlength(twooptX,twooptY)
        if optdistance < shrinkingdistance:     #keep the swap
            shrinkingdistance = optdistance
            #optplotupdate()
            #print(round(pathlength(twooptX,twooptY),2))
        else:                                   #undo the swap
            oldX = twooptX[(i+(rand))%len(twooptX)]
            oldY = twooptY[(i+(rand))%len(twooptX)]
            twooptX[(i+(rand))%len(twooptX)] = twooptX[i]
            twooptY[(i+(rand))%len(twooptX)] = twooptY[i]
            twooptX[i] = oldX
            twooptY[i] = oldY

fig = plt.figure(figsize=(12, 12), edgecolor='w')
m = Basemap(projection='cyl', resolution='l',
            llcrnrlat=20, urcrnrlat=65,
            llcrnrlon=-138, urcrnrlon=-55, )
m.etopo(scale=0.5, alpha=0.5)
m.drawstates(color='gray')
m.drawcoastlines(color='gray')
m.drawcountries(color='gray')
m.drawstates(color='gray')
for i in range(len(caps)):
    # Map (long, lat) to (x, y) for plotting
    x, y = m(alltheypoints[i], allthexpoints[i])
    plt.plot(x, y, 'ok', markersize=5)
    plt.text(x, y, ' ' + caps[i], fontsize=7);
plt.plot(twooptY,twooptX)
print("2opt Produced a path length of {} Miles".format(round(pathlength(twooptY,twooptX),2)))
```
